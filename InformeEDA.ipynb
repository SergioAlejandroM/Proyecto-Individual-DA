# Analisis Exploratorio de Datos (EDA)

```python
#Importacion de librerias
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Carga de datos
Hojas = pd.read_excel("./Datasets/Internet.xlsx", sheet_name= None)
```

```python
# Acceder a Acc_vel_loc_sinrangos
Acc_vel_loc_sinrangos =  Hojas['Acc_vel_loc_sinrangos']
display(Acc_vel_loc_sinrangos.head())
```

```python
# Obtener información general del DataFrame
print(Acc_vel_loc_sinrangos.info())

# Resumen estadístico de las columnas numéricas
display(Acc_vel_loc_sinrangos.describe())
```

```python
# Visualización de valores faltantes
plt.figure(figsize=(22, 16))
sns.heatmap(Acc_vel_loc_sinrangos.isnull(), cbar=False, cmap='viridis')
plt.title('Mapa de Calor de Valores Faltantes')
plt.show()

# Contar los valores faltantes por columna
print(Acc_vel_loc_sinrangos.isnull().sum())
```

```python
# Diagrama de caja (box plot) para cada columna de velocidad
vel_columns = Acc_vel_loc_sinrangos.columns[5:]
plt.figure(figsize=(20, 10))
sns.boxplot(data=Acc_vel_loc_sinrangos[vel_columns])
plt.xticks(rotation=90)
plt.title('Diagrama de Caja de Velocidades')
plt.show()
```

```python
# Verificar duplicados basados en todas las columnas
duplicados = Acc_vel_loc_sinrangos.duplicated()

# Mostrar las filas duplicadas
df_duplicados = Acc_vel_loc_sinrangos[duplicados]
print(df_duplicados)
```
